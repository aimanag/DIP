```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/drive/MyDrive/images (1).jpeg', 0)

# Perform Fourier Transform
fft = np.fft.fft2(image)
fft_shift = np.fft.fftshift(fft)

# Visualize the frequency spectrum
magnitude_spectrum = 20 * np.log(np.abs(fft_shift))

plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()

# Apply a high-pass filter in the frequency domain
rows, cols = image.shape
crow, ccol = rows // 2, cols // 2
fft_shift[crow - 30: crow + 30, ccol - 30: ccol + 30] = 0

# Perform Inverse Fourier Transform
fft_ishift = np.fft.ifftshift(fft_shift)
image_filtered = np.fft.ifft2(fft_ishift)
image_filtered = np.abs(image_filtered)

# Display the filtered image
plt.imshow(image_filtered, cmap='gray')
plt.title('Filtered Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
